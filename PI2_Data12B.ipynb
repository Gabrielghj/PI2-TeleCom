A continuación se crea un dataframe con los indices de inflación anual desde 2014 a 2022 provistos por el Indec

```python
data = {
    'Año': [2022, 2021, 2020, 2019, 2018, 2017, 2016, 2015, 2014],
    'Inflación': [94.79, 50.93, 36.15, 53.83, 47.65, 24.80, 34.59, 17.19, 28.27]
}

inflacion_df = pd.DataFrame(data)
```

```python
inflacion_df
```

```python
# Se eliminan las columnas  'Periodo' y 'Trimestre' del DataFrame 'ingresos'.
ingresos = ingresos.drop('Periodo', axis=1)
ingresos = ingresos.drop('Trimestre', axis=1)
ingresos
```

```python
# Se calcula el promedio de cada columna agrupado por año y se devuelve un nuevo DataFrame con los resultados,
# donde cada fila representa un año y contiene los promedios de las columnas para ese año.
ingresos = ingresos.groupby('Año').mean().reset_index()
ingresos
```

```python
# Se combinan las filas de ambos DataFrames en función de la columna 'Año'.
ingresos = ingresos.merge(inflacion_df, on=['Año'])
ingresos
```

```python
# Convertir la inflación a decimal
ingresos['Inflación'] = ingresos['Inflación'] / 100
ingresos['Inflación']
```

```python
# Calcular los ingresos reales
ingresos['Ingresos Reales (miles de pesos)'] = ingresos['Ingresos (miles de pesos)'] / (1 + ingresos['Inflación'])

# Redondear los ingresos reales a 2 decimales
ingresos['Ingresos Reales (miles de pesos)'] = ingresos['Ingresos Reales (miles de pesos)'].round(2)
```

KPI: incremento del 5% cada 100 habitantes

```python
# Se eliminan las columnas 'Periodo' y 'Trimestre'
acceso_nacional = acceso_nacional.drop('Periodo', axis=1)
acceso_nacional = acceso_nacional.drop('Trimestre', axis=1)
acceso_nacional
```

```python
# Se agrupa el DataFrame acceso_nacional por año, se calcula la media de cada grupo y se restablecen los índices.
acceso_nacional = acceso_nacional.groupby('Año').mean().reset_index()
acceso_nacional
```

```python
# Se calcula el incremento porcentual entre los valores de 'Accesos por cada 100 hab' en filas consecutivas y se guardan
# los resultados en la columna 'Incremento Porcentaje'.
acceso_nacional['Incremento Porcentaje'] = (acceso_nacional['Accesos por cada 100 hab'] - acceso_nacional['Accesos por cada 100 hab'].shift(1)) / acceso_nacional['Accesos por cada 100 hab'].shift(1)

# Se toman los valores anteriores de 'Accesos por cada 100 hab', se los multiplica por 1.05 y se asigna los resultados a la columna
#'Accesos por cada 100 hab +5%' en el DataFrame acceso_nacional
acceso_nacional['Accesos por cada 100 hab +5%'] = acceso_nacional['Accesos por cada 100 hab'].shift(1) * 1.05

# Se aplica la función lambda a cada valor de la columna para redondearlo a un decimal
acceso_nacional['Accesos por cada 100 hogares'] = acceso_nacional['Accesos por cada 100 hogares'].apply(lambda x: round(x, 1))

acceso_nacional['Accesos por cada 100 hab'] = acceso_nacional['Accesos por cada 100 hab'].apply(lambda x: round(x, 1))

# Se redondea a 4 decimales
acceso_nacional['Incremento Porcentaje'] = acceso_nacional['Incremento Porcentaje'].apply(lambda x: round(x, 4))

# Se redondea a un decimal
acceso_nacional['Accesos por cada 100 hab +5%'] = acceso_nacional['Accesos por cada 100 hab +5%'].apply(lambda x: round(x, 1))
acceso_nacional
```

```python
# Guardar el dataframe como archivo CSV
acceso_nacional.to_csv('acceso_nacional.csv', index=False)
```

KPI: Incremento de uso de la Fibra Optica del 25% de un año al siguiente

```python
# Se eliminan las columnas 'Periodo' y 'Trimestre'
acceso_tecn = acceso_tecn.drop('Periodo', axis=1)
acceso_tecn = acceso_tecn.drop('Trimestre', axis=1)
acceso_tecn
```

```python
# Se agrupa el DataFrame acceso_tecn por año, se calcula la media de cada grupo y se restablecen los índices.
acceso_tecn = acceso_tecn.groupby('Año').mean().reset_index()
acceso_tecn
```

```python
# Se calcula el porcentaje de aumento en la columna Fibra óptica de un mes a otro.
# Se toma la diferencia entre los valores de la columna Fibra óptica para dos meses consecutivos y se divide por el
# valor de la columna Fibra óptica para el mes anterior.
acceso_tecn['Incremento Porcentaje'] = (acceso_tecn['Fibra óptica'] - acceso_tecn['Fibra óptica'].shift(1)) / acceso_tecn['Fibra óptica'].shift(1)

# Se toma el valor de la columna Fibra óptica para el mes anterior y se lo multiplica por 1.25,
# que es un aumento del 25%. El resultado da la cantidad de accesos a fibra óptica que se espera que
# haya en el mes actual si el aumento es el 25% .
acceso_tecn['Fibra óptica +25%'] = acceso_tecn['Fibra óptica'].shift(1) * 1.25
acceso_tecn
```

```python
# Guardar el dataframe como archivo CSV
acceso_tecn.to_csv('acceso_tecn.csv', index=False)
```
